```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"H:\Dataset\Social_Network_Ads.csv")
```

```python
df.head()
```

```python
pd.set_option("display.max_columns",None)
pd.set_option("display.max_rows",None)
```

```python
df.drop(columns = "User ID",inplace = True)
```

```python
dummies = pd.get_dummies(df.Gender)
dummies
```

```python
dummies.drop(columns = "Female",inplace = True)
dummies
```

```python
df = pd.concat([df,dummies],axis = 1)
df
```

```python
df.drop(columns="Gender",inplace=True)
df
```

```python
X = df.drop(columns = "Purchased")
y = df["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.25,random_state=51)
```

```python
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
print("Model Accuracy -->%.2f"%(log.score(X_test,y_test)*100))
```

```python
from sklearn.metrics import mean_squared_error
y_pred = log.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```
