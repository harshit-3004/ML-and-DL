```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel(r"H:\Dataset\salary.xlsx")
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
df["company_no"] = le_company.fit_transform(df["company"])
df["job_no"] = le_job.fit_transform(df["job"])
df["degree_no"] = le_degree.fit_transform(df["degree"])
```

```python
df
```

```python
df.drop(columns = ["company","job","degree"],inplace=True)
```

```python
df
```

```python
X = df.drop(columns = ["salary_more_than_100k"])
y = df["salary_more_than_100k"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn import tree
```

```python
dt = tree.DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
dt.score(X_test,y_test)
```
